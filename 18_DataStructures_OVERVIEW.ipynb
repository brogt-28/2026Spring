````python
class Node:
   def __init__(self, data):
       self.data = data
       self.next = None

class LinkedList:
   def __init__(self):
       self.head = None
   
   def append(self, data):
       if not self.head:
           self.head = Node(data)
           return
       temp = self.head
       while temp.next:
           temp = temp.next
       temp.next = Node(data)

   def display(self):
       temp = self.head
       while temp:
           print(temp.data, end=" -> ")
           temp = temp.next

   def find(self, target):
       temp = self.head
       while temp:
           if temp.data==target:
               return True
           temp = temp.next
       return False

# VALIDATION TEST: build the list 3 4 5 6 using the above code

lyst = LinkedList()
lyst.append(3)
lyst.append(4)
lyst.append(5)
lyst.append(6)
lyst.display()
if lyst.find(6):
    print 
lyst.find(7)
````

## **Essential Data Structures Cheatsheet**

### **1. Lists (`list`)**
**Use Cases**: Storing and manipulating ordered collections of data.  
**Key Operations**:
   - Append: `list.append(x)` → O(1)  
   - Insert: `list.insert(i, x)` → O(n)  
   - Search: `x in list` → O(n)  
   - Sorting: `list.sort()` → O(n log n)  
**Example**:
```python
data = [10, 20, 30, 40]
data.append(50)  # Adds 50 at the end
data[2] = 25      # Updates value at index 2
```
🔹 **Why Important?** Used for **data preprocessing and simple storage**.  This will probably be your most common go-to form of storage.

---

### **2. Tuples (`tuple`)**
**Use Cases**: Immutable data storage (e.g., database records).  
**Example**:
```python
record = ("Alice", 25, "Analyst")
print(record[1])  # Output: 25
```
🔹 **Why Important?** **Faster than lists** (immutable & hashable), great for **fixed data**.

---

### **3. Dictionaries (`dict`)**
**Use Cases**: Fast lookups, mapping keys to values (e.g., JSON data, categorical encoding).  
**Example**:
```python
movie_ratings = {"Inception": 8.8, "Interstellar": 8.6}
print(movie_ratings["Inception"])  # Output: 8.8
```
🔹 **Why Important?** Used in **data cleaning, lookups, and JSON/XML data handling**.

---

### **4. Sets (`set`)**
 **Use Cases**: Removing duplicates, membership testing, set operations (e.g., finding unique values in datasets).  
 **Example**:
```python
ids = {101, 102, 103, 101}
print(ids)  # Output: {101, 102, 103} (removes duplicate)
```
🔹 **Why Important?** Used for **finding unique values, merging datasets, and fast lookups**.

---

### **5. Queues (`collections.deque`)**
 **Use Cases**: First-In-First-Out (FIFO) operations. Queues are lists where inserts/deletes happen at opposite ends (just like "line" where you wait you turn).  (e.g., processing logs, customer service queues).  It's a built in special class in Python (but it could be implemented using a list).
**Example**:
```python
from collections import deque
q = deque(["Alice", "Bob", "Charlie"])
q.append("David")    # Add to queue
print(q.popleft())   # Output: Alice (FIFO)
```
🔹 **Why Important?** Used in **streaming data and scheduling tasks**.

---

### **6. Stacks (Using `list` or `deque`)**
**Use Cases**: Last-In-First-Out (LIFO) operations (e.g., undo/redo, backtracking algorithms).  Inserts/Deletes happen at the same end of a list.  To do use a stack in python, most people just use a list.
**Example**:
```python
stack = []
stack.append(10)  # Push
stack.append(20)
print(stack.pop())  # Output: 20 (LIFO)
```
🔹 **Why Important?** Used in **recursion**.

---

### **7. Heaps (`heapq`)**
**Use Cases**: Efficiently retrieving the smallest or largest elements (e.g., top-N queries, priority scheduling).  
**Example**:
```python
import heapq
data = [4, 1, 7, 3, 8]
heapq.heapify(data)
print(heapq.heappop(data))  # Output: 1 (smallest element)
```
🔹 **Why Important?** Used in **finding top-N elements (e.g., top 10 customers by sales)**.

---

### **8. Trees (Using `sortedcontainers` or `networkx`)**
**Use Cases**: Hierarchical data structures (e.g., database indexing, decision trees).   **Example: Using `SortedDict` (BST-like behavior)**
```python
from sortedcontainers import SortedDict
tree = SortedDict({"Apple": 5, "Banana": 2, "Cherry": 8})
print(tree.keys())  # Sorted output: ['Apple', 'Banana', 'Cherry']
```
🔹 **Why Important?** Used in **searching, decision trees, and hierarchical data**.

---

### **9. Graphs (Using `networkx`)**
**Use Cases**: Modeling relationships between entities (e.g., social networks, recommendation systems).  
**Example**:
```python
import networkx as nx
G = nx.Graph()
G.add_edge("Alice", "Bob")
G.add_edge("Bob", "Charlie")
print(nx.shortest_path(G, "Alice", "Charlie"))
```
🔹 **Why Important?** Used in **social network analysis, fraud detection, and supply chain analytics**.

---

## **Summary: What Should Data Analysts Focus On?**
| **Data Structure** | **Why Important?** | **Example Use Cases** |
|--------------------|-------------------|-----------------------|
| **Lists** | Ordered, mutable | Data cleaning, storing rows |
| **Tuples** | Immutable | Database records, function returns |
| **Dictionaries** | Fast lookups | JSON data, mapping keys to values |
| **Sets** | Unique elements | Removing duplicates, merging datasets |
| **Queues** | FIFO processing | Log processing, customer service |
| **Stacks** | LIFO processing | Undo/redo, backtracking |
| **Heaps** | Efficient min/max retrieval | Finding top-N items |
| **Trees** | Hierarchical data | Decision trees, indexing |
| **Graphs** | Network relationships | Social networks, recommendations |
