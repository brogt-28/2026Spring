```python
 #NumPy vs Pandas 

import pandas as pd
import numpy as np

# -----------------------------
# Step 1: Creating some random Data 
# -----------------------------
data = {
    "Student": ["Ava", "Ben", "Cara"],
    "Exam1": [80, 70, 90],
    "Exam2": [85, 75, 95],
    "Exam3": [90, 80, 100]
}

df = pd.DataFrame(data)

print("Original DataFrame:")
print(df)
print()

# -----------------------------
# Step 2: Pandas task (table logic)
# -----------------------------
# Pandas is great for simple column operations
df["Average"] = (df["Exam1"] + df["Exam2"] + df["Exam3"]) / 3

print("Added Average using pandas:")
print(df)
print()

# -----------------------------
# Step 3: NumPy task (where NumPy shines)
# -----------------------------
# Suppose each exam has a DIFFERENT weight:
# Exam1 = 20%, Exam2 = 30%, Exam3 = 50%

weights = np.array([0.2, 0.3, 0.5])

# Get just the numeric scores as a NumPy array
scores = df[["Exam1", "Exam2", "Exam3"]].to_numpy()

print("Scores array:")
print(scores)
print()

# Multiply each column by its weight (broadcasting!)
weighted_scores = scores * weights

print("After applying weights (NumPy broadcasting):")
print(weighted_scores)
print()

# Sum across each row to get weighted average
weighted_average = np.sum(weighted_scores, axis=1)
#np.sum(scores)          # axis=None everything, so if it's a 2D array it's one number
#np.sum(scores, axis=0)  # down columns
#np.sum(scores, axis=1)  # across rows


print("Final weighted averages:")
print(weighted_average)
```
