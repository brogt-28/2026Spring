```python
#1. Load the dataset into a pandas DataFrame and print the header.


import pandas as pd

dance = pd.read_csv("dance_comp.csv")

dance.head()
```

```python
print(dance.shape)

dance.info()

print(dance.isnull().sum())
```

```python
import numpy as np




#Handle missing values by filling with appropriate values (NaN).

dance["Category"]= dance["Category"].fillna("Unknown")

dance["Country"]= dance["Country"].fillna("Unknown")

print(dance.isnull().sum())
```

```python
#Compute the average score per category (Genre of dance).

dance.groupby('Category')['Score'].mean().sort_values(ascending=False)
```

```python
#Find the highest-ranked dancer in each category.

# 1. Find the index 
top_indices = dance.groupby('Category')['Ranking'].idxmin()

# 2. Use those indices to pull the full rows (Dancer and Category)
highest_ranked_dancers = dance.loc[top_indices, ['Category', 'Dancer', 'Ranking']]

# 3. Sort them to see the top performers across all categories
print(highest_ranked_dancers.sort_values('Ranking'))
```

```python
#Filter results to show performances from a specific competition (e.g., "World Dance Finals").
target_comp = "World Dance Finals"
comp_data = dance[dance['Competition'] == target_comp]

comp_data.head()
```

```python
#Create a new column indicating whether the dancer placed in the top 3.


dance['Top_3'] = dance['Ranking'] <= 3

dance.head(10)
```

```python
#Calculate the average score per country.

dance.groupby('Country')['Score'].mean().sort_values(ascending=False)
```

```python
#Calculate most common winning category.


winners = dance[dance['Ranking'] == 1]

most_common_win_cat = winners['Category'].value_counts().idxmax()

print(f"The most common winning category is: {most_common_win_cat} ")

```
