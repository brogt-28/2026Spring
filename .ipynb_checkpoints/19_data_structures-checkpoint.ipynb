# Data Structures in a Nutshell

### Main goals
- Examine the most important data structures for data analysts
- Show **when** an analyst would use each one
- Include runnable code cells throughout


## Big picture: why data structures matter

A **data structure** is a way to organize data so operations on that data are efficient and natural.

For a data analyst, choosing the right data structure affects:
- clarity of code
- speed of common operations
- how naturally the code matches the problem

### Common analyst questions
- Do I need to keep order?
- Do I need fast lookup by key?
- Do I need unique values only?
- Am I processing things in arrival order?
- Do I need the smallest or largest value quickly?
- Am I storing hierarchy?
- Am I storing relationships?

### Very practical rule
The data structures most analysts will use constantly are:
- **lists**
- **dictionaries**
- **sets or tuples**

The others matter because certain problems become much easier once you recognize the right structure.

```python
# Small sample data we will reuse today

customers = ["Ava", "Ben", "Cara", "Dev", "Eli", "Finn", "Gia"]
regions = ["North", "South", "East", "West", "South", "North", "East"]
sales = [120, 90, 135, 110, 90, 200, 160]

list(zip(customers, regions, sales))  #zip creates triples in the data-- and list takes the triples
                                        #and put them into a list
```

# 1. Lists

A **list** is an ordered, mutable collection.

### Why analysts use lists
Lists are usually the first thing you reach for when:
- you want to store values in order
- you are building up results
- you are cleaning or transforming data
- you want a simple container

### Important properties
- ordered
- mutable (can change)
- can contain duplicates
- supports indexing

### Common operations
- `append(x)` → usually **O(1)**
- `insert(i, x)` → **O(n)**
- `x in my_list` → **O(n)**
- `sort()` → **O(n log n)**

```python
data = [10, 20, 30, 40]
print("Original:", data)

data.append(50)
print("After append:", data)

data[2] = 25
print("After update:", data)
```

```python
# Indexing and slicing

print("First element:", data[0])
print("Last element:", data[-1])
print("Slice data[1:4]:", data[1:4])
```

```python
# A realistic analyst example:
# collect all sales values above a threshold

high_sales = []

for amount in sales:
    if amount >= 130:
        high_sales.append(amount)

high_sales
```

```python
# Membership test in a list is linear time

print(200 in sales)
print(999 in sales)
```

```python
# Sorting a list

sorted_sales = sales.copy()
sorted_sales.sort()

print("Original sales:", sales)
print("Sorted sales:", sorted_sales)
```

### When a list is a good choice
Use a list when:
- order matters
- duplicates are allowed
- you want to gather values as you go

### When a list is not the best choice
A list may not be ideal when:
- you need very fast membership testing over and over
- you need to remove duplicates
- you need key/value lookup

That is when **sets** and **dictionaries** become better tools.

# 2. Tuples

A **tuple** is an ordered, immutable collection.  This is review!  

### Why analysts use tuples
Tuples are useful when the data should stay fixed:
- a record with fixed fields
- a function returning multiple values
- a coordinate like `(x, y)`

### Important properties
- ordered
- immutable
- can contain duplicates
- often used for fixed-size records (this is nice because it doesn't waste space)

```python
record = ("Alice", 25, "Analyst")
print(record)
print("Age:", record[1])
```

```python
# A function can return multiple values as a tuple

def summarize(values):
    smallest = min(values)
    largest = max(values)
    average = sum(values) / len(values)
    return smallest, largest, average

summary = summarize(sales)
summary
```

```python
minimum, maximum, avg = summarize(sales)
print("Minimum:", minimum)
print("Maximum:", maximum)
print("Average:", avg)
```

```python
# Tuples are immutable

point = (3, 7)
print(point)

# Uncomment to see the error:
# point[0] = 10
```

### Why a data analyst cares
A tuple is good when you want to signal:
> these values belong together and should not be changed

That makes tuples useful for:
- coordinates
- fixed records
- return values from functions

# 3. Dictionaries

A **dictionary** stores key/value pairs.

### Why analysts use dictionaries
Dictionaries are incredibly important for:
- fast lookup
- counting
- grouping
- mapping categories to values
- reading JSON-style data  (later in the course)

### Important properties
- keys map to values
- average lookup by key is about **O(1)**  THIS IS REALLY IMPORTANT!!
- keys must be hashable
- dictionaries are mutable

```python
movie_ratings = {"Inception": 8.8, "Interstellar": 8.6}
print(movie_ratings["Inception"])
```

```python
# Build a customer -> sales dictionary

customer_sales = {}

for i in range(len(customers)):
    customer_sales[customers[i]] = sales[i]

customer_sales
```

```python
# Region total sales

sales_by_region = {}

for i in range(len(regions)):
    region = regions[i]
    amount = sales[i]

    if region not in sales_by_region:
        sales_by_region[region] = 0

    sales_by_region[region] += amount

sales_by_region
```

```python
# Category counts

region_counts = {}

for region in regions:
    if region not in region_counts:
        region_counts[region] = 0
    region_counts[region] += 1

region_counts
```

```python
# Accessing items safely with get()-- but this is not efficient the big is ?

print(sales_by_region.get("North"))
print(sales_by_region.get("Central", 0))
```

### Why dictionaries matter so much
For a data analyst, dictionaries are one of the most useful Python tools because they naturally handle questions like:
- How many times does each category appear?
- What is the total for each region?
- What value goes with this key?
- How do I turn labels into numbers or categories into information?

If lists are general-purpose storage, dictionaries are often the best tool for **smart lookup**.

# 4. Sets

A **set** stores unique values.

### Why analysts use sets
Sets are great for:
- removing duplicates
- fast membership testing
- comparing two groups
- finding overlap between datasets

### Important properties
- unordered collection of unique elements
- average membership test is about **O(1)**
- duplicates are automatically removed

```python
ids = {101, 102, 103, 101}
print(ids)
```

```python
# Removing duplicates from a list

raw_regions = ["North", "South", "East", "West", "South", "North"]
unique_regions = set(raw_regions)

print("Original:", raw_regions)
print("Unique:", unique_regions)
```

```python
# Membership testing

allowed_regions = {"North", "South", "East", "West"}
print("South" in allowed_regions)
print("Central" in allowed_regions)
```

```python
# Comparing groups with set operations

survey_a = {"Ava", "Ben", "Cara", "Dev"}
survey_b = {"Cara", "Dev", "Eli", "Finn"}

print("Intersection:", survey_a & survey_b)
print("Union:", survey_a | survey_b)
print("Only in A:", survey_a - survey_b)
print("Only in B:", survey_b - survey_a)
```

### Why sets matter
A set is the right tool when your question sounds like:
- Which values are unique?
- Which IDs appear in both datasets?
- Have I seen this already?
- Which categories are missing or new?

That makes sets very practical for data cleaning and comparison.

# 5. Stacks

A **stack** is a **Last-In, First-Out (LIFO)** structure.

That means:
- the last item added is the first one removed

### Real-world analogy
Think of a stack of trays:
- add a tray to the top
- remove a tray from the top

### Why analysts should know stacks
Stacks show up in:
- undo/redo systems
- backtracking
- recursion concepts
- depth-first search
- parsing expressions

In Python, people usually use a **list** as a stack.

```python
stack = []
stack.append(10) # push
stack.append(20)
stack.append(30)

print("Stack:", stack)

item = stack.pop() # pop from the end
print("Popped:", item)
print("Stack now:", stack)
```

```python
# Another small stack example with text edits

history = []

text = ""
text += "A"
history.append(text)

text += "B"
history.append(text)

text += "C"
history.append(text)

print("Current text:", text)
print("History:", history)

history.pop()     # remove latest version
text = history[-1]
print("After undo:", text)
```

```python
# Stack behavior is LIFO "Last in first out"

stack = []
for item in ["first", "second", "third"]:
    stack.append(item)

while stack:
    print(stack.pop())
```

### Important performance note
Using a Python list as a stack is efficient when you:
- append at the end
- pop from the end

Both of those are usually **O(1)**.

### Stack the basic behavior:
- put on one end
- take off from the same end

That is the main idea to remember.

# 6. Queues

A **queue** is a **First-In, First-Out (FIFO)** structure.

That means:
- the first item added is the first one removed

### Real-world analogy
Think of a line:
- people join at the back
- people leave from the front

### Why analysts should know queues
Queues are useful for:
- jobs arriving over time
- customer support requests
- streaming data
- scheduling
- breadth-first search

In Python, the best built-in tool for a queue is usually **`collections.deque`**.

```python
from collections import deque

q = deque(["Alice", "Bob", "Charlie"])
print("Original queue:", q)

q.append("David")
print("After enqueue:", q)

first_person = q.popleft()
print("Dequeued:", first_person)
print("Queue now:", q)
```

```python
# Simulate processing support tickets in arrival order

tickets = deque()

tickets.append("ticket_101")
tickets.append("ticket_102")
tickets.append("ticket_103")

while tickets:
    current = tickets.popleft()
    print("Processing", current)
```

```python
# FIFO behavior

q = deque()
for item in ["first", "second", "third"]:
    q.append(item)

while q:
    print(q.popleft())
```

### Why not use a regular list as a queue?
You *can*, but it is not ideal.

For a list:
- `pop(0)` is **O(n)** because everything shifts left

For a deque:
- `popleft()` is **O(1)**

So when you truly want queue behavior, `deque` is usually the better choice.

```python
# Comparing list vs deque for queue-style removal

lst_queue = [1, 2, 3, 4]
removed = lst_queue.pop(0)
print("List queue removed:", removed)
print("List queue now:", lst_queue)

dq_queue = deque([1, 2, 3, 4])
removed = dq_queue.popleft()
print("Deque queue removed:", removed)
print("Deque queue now:", dq_queue)
```

# 7. Stack vs Queue

Students often mix these up, so let's compare them directly.

- **Stack** = LIFO = last in, first out
- **Queue** = FIFO = first in, first out

```python
from collections import deque

items = [1, 2, 3, 4]

# stack version
stack = []
for x in items:
    stack.append(x)

stack_output = []
while stack:
    stack_output.append(stack.pop())

# queue version
queue = deque()
for x in items:
    queue.append(x)

queue_output = []
while queue:
    queue_output.append(queue.popleft())

print("Original order:      ", items)
print("Stack removal order: ", stack_output)
print("Queue removal order: ", queue_output)
```

### Quick memory trick
- **stack**: think **stack of plates**
- **queue**: think **queue/line of people**

# 8. Heaps

A **heap** is a structure designed for efficient access to the smallest item
(or largest, with a variation).

Python's `heapq` module gives us a **min-heap**.

### Why analysts use heaps
Heaps are useful for:
- top-N problems
- priority queues
- keeping track of largest/smallest values efficiently
- handling streaming data without sorting everything again and again

### Important idea
A heap is **not** the same thing as a sorted list.

A min-heap guarantees:
- the smallest item is easy to access

But it does **not** mean the whole structure is sorted in order.

```python
import heapq

data = [4, 1, 7, 3, 8]
heapq.heapify(data)

print("Internal heap representation:", data)

smallest = heapq.heappop(data)
print("Smallest removed:", smallest)
print("Heap after pop:", data)
```

```python
# Top 3 largest sales values

heapq.nlargest(3, sales)
```

```python
# Top 3 smallest sales values

heapq.nsmallest(3, sales)
```

```python
# Priority queue example
# smaller priority number means more urgent

tasks = []

heapq.heappush(tasks, (2, "reply to regular email"))
heapq.heappush(tasks, (1, "fix broken dashboard"))
heapq.heappush(tasks, (3, "archive old files"))

while tasks:
    priority, task = heapq.heappop(tasks)
    print("Do next:", task, "| priority =", priority)
```

### Why a heap matters to an analyst
Suppose records keep arriving and you always want the top 10 values.
You could sort everything each time, but that may be wasteful.

A heap gives a cleaner way to maintain "most important next" items.

# 9. Trees

A **tree** is a hierarchical data structure made of nodes.

### Vocabulary
- **root**: the top node
- **parent**: a node above another
- **child**: a node below another
- **leaf**: a node with no children

### Why analysts use tree ideas
Even if analysts do not manually implement trees often, the concept matters for:
- decision trees in machine learning
- hierarchical categories
- organization charts
- file systems
- drill-down reporting

We will build a tiny binary tree manually.

```python
class TreeNode:
    def __init__(self, value):
        self.value = value
        self.left = None
        self.right = None

root = TreeNode("Sales")
root.left = TreeNode("North")
root.right = TreeNode("South")
root.left.left = TreeNode("Ava")
root.left.right = TreeNode("Finn")
root.right.left = TreeNode("Ben")
root.right.right = TreeNode("Dev")
```

```python
def inorder(node):
    if node is not None:
        inorder(node.left)
        print(node.value)
        inorder(node.right)

inorder(root)
```

```python
def preorder(node):
    if node is not None:
        print(node.value)
        preorder(node.left)
        preorder(node.right)

preorder(root)
```

```python
def postorder(node):
    if node is not None:
        postorder(node.left)
        postorder(node.right)
        print(node.value)

postorder(root)
```

You have already seen trees, so the most important reminder here is:
- trees model **hierarchy**
- traversals visit the nodes in different orders
- tree thinking appears often in analytics and machine learning

```python
# Very simple text drawing of the tree

print("          Sales")
print("         /     \\")
print("     North     South")
print("     /   \\     /   \\")
print("   Ava  Finn  Ben   Dev")
```

# 10. Graphs

A **graph** models relationships.

A graph has:
- **nodes** (or vertices)
- **edges** connecting nodes

### Why analysts use graphs
Graphs are useful for:
- social networks
- recommendation systems
- fraud detection
- transportation networks
- supply chains
- web links

We will keep this very simple.

```python
# Graph represented as a dictionary of lists (adjacency list)

graph = {
    "Alice": ["Bob", "Cara"],
    "Bob": ["Alice", "Dev"],
    "Cara": ["Alice", "Dev"],
    "Dev": ["Bob", "Cara", "Eli"],
    "Eli": ["Dev"]
}

graph
```

```python
# Breadth-first search (BFS) using a queue

from collections import deque

def bfs(graph, start):
    visited = set()
    q = deque([start])

    while q:
        node = q.popleft()
        if node not in visited:
            visited.add(node)
            for neighbor in graph[node]:
                if neighbor not in visited:
                    q.append(neighbor)
    return visited

bfs(graph, "Alice")
```

```python
# Shortest path in an unweighted graph

def shortest_path(graph, start, goal):
    q = deque([[start]])
    visited = set()

    while q:
        path = q.popleft()
        node = path[-1]

        if node == goal:
            return path

        if node not in visited:
            visited.add(node)
            for neighbor in graph[node]:
                q.append(path + [neighbor])

print(shortest_path(graph, "Alice", "Eli"))
```

### Why graphs matter
Use graph thinking when your question is about **relationships**:
- who is connected to whom?
- what is the shortest route?
- what accounts interact with each other?
- how do items relate in a network?

That is different from a simple table or list.

```python
# Optional simple picture with networkx
# This cell will still run even if networkx is not installed.

import matplotlib.pyplot as plt

try:
    import networkx as nx

    G = nx.Graph()
    G.add_edge("Alice", "Bob")
    G.add_edge("Alice", "Cara")
    G.add_edge("Bob", "Dev")
    G.add_edge("Cara", "Dev")
    G.add_edge("Dev", "Eli")

    plt.figure(figsize=(6, 4))
    nx.draw(G, with_labels=True)
    plt.title("Simple Relationship Graph")
    plt.show()
except ImportError:
    print("networkx is not installed, so this graph picture is skipped.")
```

# 12. Complexity summary

These are rough, common average-case ideas.

```python
summary = [
    ("list", "ordered mutable storage", "append O(1), search O(n)"),
    ("tuple", "fixed ordered record", "index access O(1)"),
    ("dict", "key/value lookup", "lookup O(1) average"),
    ("set", "unique values / membership", "membership O(1) average"),
    ("stack", "LIFO", "push/pop end O(1)"),
    ("queue", "FIFO", "append/popleft O(1) with deque"),
    ("heap", "smallest/largest efficiently", "push/pop O(log n)"),
    ("tree", "hierarchy", "depends on tree type"),
    ("graph", "relationships", "depends on algorithm")
]

for row in summary:
    print(row)
```

## Summary table

| Data Structure | Main Strength | Typical Analyst Use |
|---|---|---|
| `list` | Ordered, mutable | store values, clean data, collect results |
| `tuple` | Fixed, immutable | records, return values |
| `dict` | Fast lookup by key | counts, grouping, mappings |
| `set` | Unique values | remove duplicates, overlap checks |
| stack | LIFO behavior | undo, recursion ideas, DFS |
| queue | FIFO behavior | arrivals, scheduling, BFS |
| heap | efficient top/min access | top-N, priority work |
| tree | hierarchy | decision trees, categories |
| graph | relationships | networks, shortest paths |
