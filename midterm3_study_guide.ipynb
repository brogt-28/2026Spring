```python
# Harder Problem 3
# We are modeling connections between cities.
# Each city is connected to other cities it shares events with.

# The graph is represented as a dictionary:
graph = {
    "Memphis": ["Nashville", "Jackson"],
    "Nashville": ["Memphis", "Knoxville"],
    "Knoxville": ["Nashville"],
    "Jackson": ["Memphis"]
}

# 1. How many neighbors does Memphis have?
# 2. Write code to print all neighbors of Nashville.
# 3. Write code to check if Memphis is connected to Knoxville.
# 4. Write code to count how many total connections exist in the graph.
#    (Hint: be careful not to double count edges)


memphis_neighbors_count = len(graph["Memphis"])
print(f"1. Memphis has {memphis_neighbors_count} neighbors.")

# 2. Write code to print all neighbors of Nashville.
print("2. Neighbors of Nashville:", graph["Nashville"])

# 3. Write code to check if Memphis is connected to Knoxville.
# We check if "Knoxville" is in the list of Memphis's neighbors.
is_connected = "Knoxville" in graph["Memphis"]
print(f"3. Is Memphis connected to Knoxville? {is_connected}")

# 4. Write code to count how many total connections exist in the graph.
# (Hint: be careful not to double count edges)
total_connections = 0
for city in graph:
    total_connections += len(graph[city])

# Since every connection is listed twice (A->B and B->A), divide by 2
final_count = total_connections // 2
print(f"4. Total unique connections: {final_count}")
```
