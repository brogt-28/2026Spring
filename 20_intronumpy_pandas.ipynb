
# Part 1 — NumPy, introduced clearly

Before NumPy, think about a regular Python list.
A list is flexible. It can hold many kinds of values, and it is great for general programming.

But when we want to do **numerical computing**, lists are clunky:
- math on every item takes more code
- lists are not designed for fast large-scale numerical operations

NumPy gives us a better tool for numbers:
- a **NumPy array** stores numerical data efficiently
- it lets us do math on an entire collection at once


```python

import numpy as np

scores_list = [72, 88, 91, 84, 95]

print("Python list:")
print(scores_list)
print("Type:", type(scores_list))
```

```python

# A list does not automatically do element-by-element math
# This repeats the list rather than doubling each value
print(scores_list * 2)
```

```python

scores = np.array(scores_list)

print("NumPy array:")
print(scores)
print("Type:", type(scores))
print("Shape:", scores.shape)
```

```python

# NumPy does element-by-element math
print("scores + 5:", scores + 5)
print("scores * 2:", scores * 2)
print("Average:", np.mean(scores))
print("Highest score:", np.max(scores))
```


## Why this matters

With NumPy, we can think in terms of:
- "add 5 to every score"
- "double every value"
- "find the average"

instead of writing loops for every small task.

That is why NumPy is the foundation for a lot of scientific and data-analysis work in Python.



## Example: city coordinates

Suppose each city has:
- a latitude
- a longitude

We can store all coordinates in a 2D NumPy array and compute distances from a reference point.


```python

city_names = np.array(["Memphis", "Nashville", "St. Louis", "Dallas", "Chicago", "Atlanta"])

# Each row is [latitude, longitude]
city_coords = np.array([
    [35.1495, -90.0490],
    [36.1627, -86.7816],
    [38.6270, -90.1994],
    [32.7767, -96.7970],
    [41.8781, -87.6298],
    [33.7490, -84.3880]
])

center = np.array([39.8283, -98.5795])

print("City coordinates array:")
print(city_coords)
print("Shape:", city_coords.shape)
```

```python

# Compute Euclidean distance from each city to the reference point
distances = np.sqrt(np.sum((city_coords - center) ** 2, axis=1))

print("Distances from the reference point:")
for i in range(len(city_names)):
    print(city_names[i], "->", round(distances[i], 2))
```

```python

# Find the city closest to the reference point
closest_index = np.argmin(distances)
print("Closest city:", city_names[closest_index])
```
